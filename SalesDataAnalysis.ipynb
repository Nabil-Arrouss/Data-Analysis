```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

#### concatenating the 12 data files in 1 csv file

```python
# list comprehension to iterate through the Sales_Data directory 
files = [file for file in os.listdir('./Sales_Data')]

# create a dataframe to store all the 12 files
all_data = pd.DataFrame()

# read the data files, concatenate them and store them in the dataframe 
for file in files:
    data = pd.read_csv("./Sales_Data/"+file)
    all_data = pd.concat([all_data, data])

# save the dataframe as csv 
all_data.to_csv("all_data.csv", index=False)
```

#### Reading the csv file data 

```python
all_data = pd.read_csv("all_data.csv")
all_data.head(50)
```

#### Cleaning the data!

```python
# printing 50 samples
all_data.head(50)
```

```python
# drop samples (rows) of NaN
all_data = all_data.dropna(how='all')
all_data.head()
```

```python
# find the error (or) and delete them
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

```python
# convert columns to correct types

all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data.shape
```

```python
all_data.head()
```

### Q1) What was the best month for sales? How much was earned that month? 

```python
# add 'Month' column to the dataframe 

all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')

all_data.head(100)
```

```python
# add 'Sales' column and calculate it 
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
# group the rows based on the 'Month' column and sum its corresponding numerical values, then select the 'Sales' column

all_data.groupby('Month').sum()['Sales']
```

### A1) The best month for sales is 12 (Decemeber) and 4613443.34 was earned on this month

```python
# save the results 
results = all_data.groupby('Month').sum()

months = range(1,13)

# plot the resuls
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Months')
plt.show()
```

### Q2) what city had the highest number of sales?

```python
all_data.head()
```

```python
# using the .apply function to create a column called 'City' by spliting the 'Purchase Address' (917 1st St, Dallas, TX 75001)

# get_city function splits the address using the ',' as a delimeter.
def get_city(address):
    return address.split(',')[1]

# get_state function splits the address 2 times, using the ',' and ' ' as a delimeter
def get_state(address):
    return address.split(',')[2].split(' ')[1]

# create a column called 'City' and apply the functions to get city and state name using .apply method 
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + get_state(x))

all_data.head()
```

```python
# group by the cities and sum the nemurical values, then select 'Sales' column to see which city has the highest sales
cities_results = all_data.groupby('City').sum()['Sales']

# get the city name and the highest sales
city = cities_results.idxmax()
sales = cities_results.max()

# displaying the results of all cities
print(f"{cities_results}\n")
# print the city name with the highest sales
print(f"{city} is the highest city with {sales} $ of sales")
```

```python
# extract all unique cities and save them after grouping them 
cities = [city for city, df in all_data.groupby('City')]

# plot the results 
plt.bar(cities, cities_results)
plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City Name')
plt.show()
```

### A1) The city with the highest number of sales is San Francisco CA with 8262203.91 earned 

### Q3) What time should we display ads to maximize likelihood of customer's buying products? 

```python
all_data.head()
```

```python
# convert the 'Order Date' column to a date/time object 
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data.head()
```

```python
# create 'Hour' and 'Minute' columns and extract them from the 'order date' column
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
# extract hours after grouping them 
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.grid()
plt.show()
```

### A3) The time recommended to display ads is around 11am or 7pm (according to the plot)

### Q4) What products are most often sold together?

```python
# group by the 'Order ID' using the 'duplicated' method
df = all_data[all_data['Order ID'].duplicated(keep=False)]

# create a column with all products grouped by ID
df['Grouped Products'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

# drop all duplicates
df = df[['Order ID', 'Grouped Products']].drop_duplicates()
df.head()
```

```python
# import libraries
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in df['Grouped Products']:
    row_list = row.split(',')
    # 
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)
```

### A5) The most sold item is "AAA Batteries (4-pack)"

```python
# Get the price of each product and plot them as secondary y-axis on the plot above
prices = all_data.groupby('Product')['Price Each'].first()
print(prices)

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)
plt.show()
```
